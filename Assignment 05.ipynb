Create a list called numbers containing at least 8 integers.
Using a for loop, do the following:
o	Print all numbers in the list.
o	Print the sum of all numbers.
o	Print only the even numbers.


```python
numbers = [5, 12, 7, 20, 3, 15, 9, 18]
print("All numbers in the list:")
for num in numbers:
    print(num)
```

```python
total = sum(numbers)
print("Sum of all numbers:")
print(total)
```

```python
print("Even numbers in the list:")
for num in numbers:
    if num % 2 == 0:
        print(num)
```

1.	Create a dictionary called student with the following keys:
o	name
o	age
o	course
o	marks (must store a list of 3 marks)
2.	Using loops:
o	Print all keys and their values.
o	Calculate the average of the marks.
o	Display whether the student Passed (average ≥ 50) or Failed.


```python
student = {
    "name": "Ammarah",
    "age": 20,
    "course": "Mathematics",
    "marks": [65, 70, 55]
}
print(student)
```

```python
print("Student Details:")
for key, value in student.items():
    print(key, ":", value)
```

```python
total = 0
for mark in student["marks"]:
    total += mark

print(total)
```

```python
average = total / len(student["marks"])
print(average)
```

```python
if average >= 50:
    print("Passed")
else:
    print("Failed")
```

1.	Create a list called employees.
2.	Each employee must be stored as a dictionary containing:
o	name
o	department
o	salary
3.	Using loops:
o	Print details of all employees.
o	Find the employee with the highest salary.
o	Calculate the total salary expense.

```python
employees = [
    {"name": "Ali", "department": "HR", "salary": 50000},
    {"name": "Sara", "department": "IT", "salary": 70000},
    {"name": "Ahmed", "department": "Finance", "salary": 60000}
]
print("Employee Details:")
for emp in employees:
    print(emp)
```

```python
highest = employees[0]
for emp in employees:
    if emp["salary"] > highest["salary"]:
        highest = emp
print("Highest Salary Employee:")
print(highest)
```

```python
total_salary = 0
for emp in employees:
    total_salary += emp["salary"]
print("Total Salary Expense:")
print(total_salary)
```

1.	Ask the user to enter numbers.
2.	Store the numbers in a list.
3.	Stop when the user enters -1.
4.	After the loop ends:
o	Print the complete list.
o	Print the largest number.
o	Print the smallest number.

```python
numbers = []
while True:
    num = int(input("Enter a number (-1 to stop): "))
    if num == -1:
        break
    numbers.append(num)
print(numbers)
```

```python
print("Largest number:")
print(max(numbers))
```

```python
print("Smallest number:")
print(min(numbers))
```

1.	Ask the user to enter a sentence.
2.	Convert the sentence into a list of words.
3.	Use a dictionary to count how many times each word appears.
4.	Print the word frequency clearly.
Example Format (for understanding only):
apple: 3
banana: 2
orange: 1


```python
sentence = input("Enter a sentence: ")
words = sentence.split()
print(words)
```

```python
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print(frequency)
```

```python
print("Word Frequency:")
for word, count in frequency.items():
    print(word, ":", count)
```

Using nested loops, print a multiplication table from 1 to 5.


```python

numbers = [1, 2, 3, 4, 5]

table_rows = []
table_dict = {}

for i in numbers:          
    row = []
    for j in numbers:      
        row.append(i * j)
    table_rows.append(row)
    table_dict[i] = row

row_sum_dict = {i: sum(row) for i, row in table_dict.items()}

print("Multiplication Table (1 to 5):")
for row in table_rows:
    print(*row) 

print("\nSum of each row:")
for row_num, total in row_sum_dict.items():
    print(f"Row {row_num} sum: {total}")
```
